## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple IF-statement (like "if clicks drop by 20%, flag it") is reactive—it only flags a page *after* the damage is already done.

Machine Learning beats a fixed rule because it can look at complex, non-linear patterns across multiple features (such as CTR drops, impression changes, and seasonality) to *predict* which pages are highly likely to decline in the future *before* the traffic actually plummets. This lets our content team act proactively rather than reactively.
